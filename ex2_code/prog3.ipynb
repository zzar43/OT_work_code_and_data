```julia
@everywhere function make_model(kappa,v0,p,Nx,Ny)
    c = v0.* ones(Nx, Ny)
    for j = 1:Ny
        for i = 1:Nx
            c[i,j] = c[i,j] + kappa*(i-1)
        end
    end
    c[p:end,:] = c[p:end,:] .+ 0.2
    return c
end

@everywhere function make_model_1(kappa,v0,p,Nx,Ny)
    c = v0.* ones(Nx, Ny)
    for j = 1:Ny
        for i = 1:Nx
            c[i,j] = c[i,j] + kappa*(i-1)
        end
    end
    return c
end
```

```julia
@everywhere begin
    Nx = 101
    Ny = 101
    c_true = make_model(0.005, 1, 51, Nx, Ny)
    rho = ones(Nx,Ny)
    h = 0.01

    # time
    Fs = 500;
    dt = 1/Fs;
    Nt = 1000;
    t = range(0,length=Nt,step=dt);
    println("CFL: ", maximum(c_true) * dt / h);

    # source
    source = source_ricker(6,0.2,t)
    source_num = 1
    source_position = zeros(Int,source_num,2)
    for i = 1:source_num
            source_position[i,:] = [5 51]
    end
    source = repeat(source, 1, 1);

    # receiver
    receiver_num = 11
    receiver_position = zeros(Int,receiver_num,2)
    for i = 1:receiver_num
        receiver_position[i,:] = [1, (i-1)*10+1]
    end

    # PML
    pml_len = 50
    pml_coef = 50;
    
end
plot(t, source[:,1])
```

## Iteration

```julia
epss = 1e-4
lambda_mix = 1e-8
M = cost_matrix_1d(t, t; p=2)
iteration_number = 500
k_normalize = 2e4

reg = 1e-4
reg_m = 1e0
```

```julia
# compute the wavefield
print("total c: ")
println(n_kappa)
@time @sync @distributed for i = 1:n_kappa
    for j = 1:n_v0
        cc = make_model(kappa[i], v0[j], p, Nx, Ny)
        data = acoustic_solver_no_wavefield(cc, rho, Nx, h, Ny, h, Nt, dt, source, source_position, receiver_position; pml_len=pml_len, pml_coef=pml_coef);
        
        cc1 = make_model_1(kappa[i], v0[j], p, Nx, Ny)
        data1 = acoustic_solver_no_wavefield(cc1, rho, Nx, h, Ny, h, Nt, dt, source, source_position, receiver_position; pml_len=pml_len, pml_coef=pml_coef);
          
        data = data - data1
        
        file_name = "wavefield3/wavefield_" * string(i) * "_" * string(j) * ".jld2"
        @save file_name data
    end
    println(i)
end
```

```julia
# compute the l2 misfit
print("total c: ")
println(n_kappa)
misfit_l2 = SharedArray{Float64}(n_kappa, n_v0)

@time @sync @distributed for i = 1:n_kappa
    for j = 1:n_v0
        file_name = "wavefield3/wavefield_" * string(i) * "_" * string(j) * ".jld2"

        data = load(file_name, "data");
        misfit_l2[i,j] = 0.5*norm(received_data-data,2)^2
    end
end
misfit_l2 = Array(misfit_l2);
```

```julia
imshow(misfit_l2); colorbar()
```

```julia
using3D()
xgrid = repeat(kappa,1,n_v0)
ygrid = repeat(v0',n_kappa,1);
```

```julia
fig = figure("pyplot_surfaceplot",figsize=(6,6))

ax = fig.add_subplot(1,1,1,projection="3d")
plot_surface(xgrid, ygrid, misfit_l2, rstride=2, edgecolors="k", cstride=2, cmap=ColorMap("viridis"), alpha=0.7, linewidth=0.25)
ylabel("z"); xlabel(L"$\\delta c\$"); zlabel("misfit function"); title("(a) \$L^2\$ distance")
ax.view_init(45, 45)

tight_layout()
```

```julia
# compute the uot misfit
print("total c: ")
println(n_kappa)
misfit_uot = SharedArray{Float64}(n_kappa, n_v0)

# @time @sync @distributed for i = 1:1
@time @sync @distributed for i = 1:n_kappa
    for j = 1:n_v0
        file_name = "wavefield3/wavefield_" * string(i) * "_" * string(j) * ".jld2"
        data = load(file_name, "data");
        
        dist1, adj = adj_source_ot_exp(data, received_data, M; reg=reg, reg_m=reg_m, iterMax=iteration_number, k=k_normalize);
        adj = nothing
        
        misfit_uot[i,j] = dist1
    end
    println(i)
end
misfit_uot = Array(misfit_uot);
```
